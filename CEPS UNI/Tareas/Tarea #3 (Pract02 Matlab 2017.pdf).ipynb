````octave
clear all, clc;
disp('Solución del ejercicio 1:')
x=-2:0.1:3;
y=(x.^2+2)./(x+5);
plot(x,y,'r')
````

````octave
clear all, clc;
disp('Solución del ejercicio 2:')
x=-1.5:0.1:1.5;
y=2.*sin(x).^2.*cos(x).^3;
z=exp(x)-2*x-3;
subplot(2,1,1),plot(x,y,'r')
subplot(2,1,2),plot(x,z,'r')
````

````octave
clear all, clc;
disp('Solución del ejercicio 3:')
x = input('Ingrese un número entero: ')
for i=1:12
    r = i*x;
    fprintf('%d * %d = %d\n',i, x, r)
end
````
