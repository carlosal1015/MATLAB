# Plotear gráficas en Matlab o GNU Octave #

Formato de los argumentos:

```Matlab
Estilo de línea:

‘-’	Use solid lines (default).
‘--’	Use dashed lines.
‘:’	Use dotted lines.
‘-.’	Use dash-dotted lines.

Marcador:

‘+’	crosshair
‘o’	circle
‘*’	star
‘.’	point
‘x’	cross
‘s’	square
‘d’	diamond
‘^’	upward-facing triangle
‘v’	downward-facing triangle
‘>’	right-facing triangle
‘<’	left-facing triangle
‘p’	pentagram
‘h’	hexagram

Colores:
‘k’	blacK
‘r’	Red
‘g’	Green
‘b’	Blue

‘c’	Cyan
‘w’	White
```

## Ejemplo ##

````octave
x = linspace(-pi, pi, 800); % Intervalo en x con 800 puntos de muestra.
y = sin(x); % Puntos de muestra.
%plot(x,y,'k-.') % Diseño 1
plot(x,y,'go') % Diseño 2
grid
````

## Ejemplo 2 ##

````octave
subplot(1, 1, 1)
x = linspace(-pi, pi, 250);
y = sin(x);
subplot(2,2,1), plot(x,y,'r')
subplot(2,2,2), plot(x,y,'m*')
subplot(2,2,3), plot(x,y,'k-.')
subplot(2,2,4), plot(x,y,'go')
````

## Ejemplo 3 ##

````octave
subplot(1,1,1)
x=linspace(-pi,pi,250);
y=sin(x);
subplot(2,1,1), plot(x,y,'r')
subplot(2,1,2), plot(x,y,'m*')
````

## Ejemplo 4 ##

````octave
subplot(1,1,1)
x=linspace(-pi,pi,250);
y=sin(x);
subplot(1,2,1), plot(x,y,'k-.')
subplot(1,2,2), plot(x,y,'go')
%~ alt + 126
````

## Ejemplo 5 ##

````octave
clear all, clc;
% Entrada de datos
fx=input('Frecuencia en x: '); %pi/4
dx=input('Fase en x: ');       %0.05
fy=input('Frecuencia en y: '); %pi/2 %Paso
dy=input('Fase en y: ');       %0.03 %Paso
% Proceso
t = 0:pi/100:2*pi;
x = cos(fx*t + dx);
y = sin(fy*t + dy);
% Salida de resultados
plot(x,y,'m*')
````
